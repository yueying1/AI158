```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from seaborn import histplot, heatmap, violinplot
```

```python
# 1.实战训练实现球队评价数据的所有数值列的相关度热力图
# 获取球队全部数值数据，绘制热力图
df = pd.read_csv('../data/team_attributes.csv')
# dfpa.dtypes
df_num = df.select_dtypes(include=['number'])
heatmap(df_num.corr(), cmap='Oranges', center=0, annot=True)
```

```python
# 实战训练根据惯用脚分类绘制评分的小提琴图
# 加载球员评价数据
df = pd.read_csv('../data/player_attributes.csv')
df.head(2)
# 设置惯用脚preferred_foot为索引
df.set_index('preferred_foot', inplace=True)
# 绘制整体评分overall_rating和潜力值potential的小提琴图
violinplot(x=df.overall_rating)
plt.show()
violinplot(x=df.potential)
plt.show()
# 根据惯用脚分类，绘制评分与潜能的小提琴图
violinplot(y=df.index, x=df.overall_rating)
plt.show()
violinplot(y=df.index, x=df.potential)
plt.show()
```
