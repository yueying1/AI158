```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#B组失序分析按年统计购买数量
df=pd.read_csv("../data/处理后数据.csv")
df=df.query("购买时年龄.notna()")
df.购买日期=pd.to_datetime(df.购买日期)
df=df.assign(
    年=df.购买日期.dt.year,
    月=df.购买日期.dt.month,
    日=df.购买日期.dt.day,
    季度=df.购买日期.dt.quarter,
    周=df.购买日期.dt.isocalendar().week
)
df
```

```python
# 按年统计购买数量
df.groupby('年').购买数量.sum().plot.bar()
plt.show()
# 按季度对比不同年的购买数量
df.groupby(['年','季度']).购买数量.sum().plot.bar()
plt.show()
# 按月统计购买数量
df.groupby(['年','月']).购买数量.sum().plot.bar()
plt.show()
# 按月对比不同年的购买数量
df.groupby(['年','月']).购买数量.sum().unstack(0).plot.bar()
plt.show()
# 按周统计购买数量
df.groupby(['年','周']).购买数量.sum().plot.bar()
plt.show()
# 按日统计购买数量
df.groupby(['年','日']).购买数量.sum().plot.bar()
plt.show()
df.groupby('购买日期').购买数量.sum().sort_valuces(ascending=False).plot.bar()
```

```python
#B组多变量分析
# 一级分类的购买量分析
df.groupby('一级分类').购买数量.sum().plot.bar()
plt.show()
# 一级分类下的二级分类数量统计
df.groupby('一级分类').二级分类.nunique().plot.bar()
plt.show()
# 一级分类下的二级分类购买量统计
#实现一级分类下的各二级分类平均购买量统计
(df.groupby('一级分类').购买数量.sum()/df.groupby('一级分类').二级分类.nunique()).plot.bar()
plt.show()
```

```python
# 按性别实现一级类别购买量分布统计
# 实现购买商品记录中婴儿性别的占比饼图
df.query('性别<2').groupby('性别').订单号.count().plot.pie(autopct='%.2f%%')
plt.show()
# 实现男婴与女婴的一级类别购买量分布饼图
a=df.query('性别<2')
a.groupby(['一级分类','性别']).购买数量.sum().unstack().rename(columns={0:'女',1:'男'}).plot.pie(autopct='%.2f%%',subplots=True,legend=False)

# 一级类别下不同性别、年龄段的购买数量分析
b=a.groupby(['一级分类','性别',df.购买时年龄.round(0)]).购买数量.sum().unstack(1).rename(columns={0:'女',1:'男'})
for i in b.index.get_level_values(0).unique():
    c=b.xs(key=i,level=0).plot.line()
```

```python
#多变量分析
df.购买日期=pd.to_datetime(df.购买日期)
# 分别统计不同性别所属用户数和购买量总和的占比
df.query('性别<2').groupby('性别').agg({'用户编号':'nunique','购买数量':'sum'}).rename(columns={0:'女',1:'男'}).plot.pie(autopct='%.2f%%',subplots=True,legend=False)
# 不同性别用户在一级分类上的购买量总和对比
df.query('性别<2').groupby(['性别','一级分类']).购买数量.sum().unstack(0).rename(columns={0:'女',1:'男'}).plot.bar()
# 一级分类购买量与累计比例的统计
a=df.groupby('一级分类').购买数量.sum().to_frame().sort_values('购买数量',ascending=False)
a.assign(累计销量=a.购买数量.cumsum()).eval('累计比例=累计销量/购买数量.sum()*100')
```
