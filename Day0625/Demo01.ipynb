```python
import requests
from lxml import etree
import time
```

```python
start = 0
h = {
    'user-agent':
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/149.0.0.0 Safari/537.36 Edg/149.0.0.0'
}
na_c = []
na_e = []
评分 = []
评分人数 = []
其他 = []

while start < 250:
    url = f'https://www.douban.com/doulist/3936288/?start={start}&sort=time&playable=0&sub_type='
    resp = requests.get(url, headers=h)
    tree = etree.HTML(resp.text)

    names = tree.xpath('//div[@class="title"]/a/text()')
    for i in names:
        name = i.split()
        if name:
            na_c.append(name[0])
            na_e.append(' '.join(name[1:]))
    评分.extend(tree.xpath('//div[@class="rating"]/span[2]/text()'))
    评分人数.extend(tree.xpath('//div[@class="rating"]/span[3]/text()'))
    其他.extend(tree.xpath('//div[@class="abstract"]/text()'))
    start += 25
    time.sleep(3)
评分人数 = [int(i[1:-4]) for i in 评分人数]
# na_c
# na_e
len(na_c), len(na_e), len(评分), len(评分人数), len(其他)
```

```python
list1 = []
for i in range(0, len(其他) - 4, 2):
    list2 = 其他[i: i + 5]
    list3 = []
    for j in list2:
        list3.append(j.strip())
    info = {}
    for j in list3:
        if ':' in j:
            info[j.split(':')[0]] = j.split(':')[1].strip()
    list1.append(info)
list1
```

```python
import csv

with open('douban.csv', mode='w', encoding='utf-8', newline='') as f:
    file = csv.writer(f)
    file.writerow(['电影中文名', '电影外文名', '评分', '评分人数', '导演', '主演', '类型', '制片国家/地区', '年份'])
    n = min(len(na_c), len(na_e), len(评分), len(评分人数), len(list1))
    for i in range(n):
        info = list1[i]
        file.writerow([na_c[i], na_e[i], 评分[i], 评分人数[i],
                       info.get('导演', ''), info.get('主演', ''),
                       info.get('类型', ''), info.get('制片国家/地区', ''), info.get('年份', '')])
```
