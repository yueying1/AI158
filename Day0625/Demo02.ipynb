```python
import requests
from lxml import etree
```

```python
url = 'https://miaoso.com/bus/beijing/'
res = requests.get(url)
print(res.status_code)
print(res)
tree = etree.HTML(res.text)
```

```python
href = tree.xpath('//div[@class="gj01_item gj01_item_01"]/ul/li[2]/a/@href') [0]
print('https://miaoso.com/bus/beijing/' + href)
newHtml = 'https://miaoso.com/bus/beijing/' + href
res = requests.get(newHtml)
tree = etree.HTML(res.text)
hrefs = tree.xpath('//div[@class="list"][2]/ul[@class="f-cb"]/li/a/@href')
titles = tree.xpath('//div[@class="list"][2]/ul[@class="f-cb"]/li/a/@title')
print(hrefs)
print(titles)
```
