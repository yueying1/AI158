```python
df = pd.read_csv('../data/team_attributes.csv', index_col='id')
# df
# ➢获得id=a3数据的buildUpPlaySpeed
#   ✓分别使用先获得行再获得列、先获得列再获得行的方式
df.loc[3, 'buildUpPlaySpeed']
df.buildUpPlaySpeed.loc[3]
# ➢获得a3、a5数据中的buildUpPlaySpeed和buildUpPlaySpeedClass
#   ✓分别使用先获得行再获得列、先获得列再获得行的方式
df.loc[[3, 5], ['buildUpPlaySpeed', 'buildUpPlaySpeedClass']]
df[['buildUpPlaySpeed', 'buildUpPlaySpeedClass']].loc[[3, 5]]
# ➢使用行列混合切片获得a3到a5数据的buildUpPlaySpeed和buildUpPlaySpeedClass
df.loc[3:5, 'buildUpPlaySpeed':'buildUpPlaySpeedClass']
# ➢修改所有数据中buildUpPlayDribbling为NAN的数据设置为0
df.loc[df['buildUpPlayDribbling'] == np.NAN] = 0
# ➢修改所有数据中buildUpPlaySpeed<30的数据buildUpPlaySpeed值为30
df.loc[df['buildUpPlaySpeed'] < 30] = 30
```

```python
df = pd.read_csv('../data/team_attributes.csv')
# 实现buildUpPlaySpeed、buildUpPlayDribbling数据列的均值
df[['buildUpPlaySpeed', 'buildUpPlayDribbling']].mean()
# 统计显示buildUpPlaySpeed与buildUpPlayDribbling、buildUpPlayPassing的相关度
df[['buildUpPlaySpeed', 'buildUpPlayDribbling', 'buildUpPlayPassing']].corr()
# 列出buildUpPlaySpeed低于平均分的数据并按照从低到高排序
df[df['buildUpPlaySpeed'] < df['buildUpPlaySpeed'].mean()].sort_values('buildUpPlaySpeed')
# 统计buildUpPlaySpeedClass的唯一值数量(不重复元素的个数)与唯一值元素（不重复元素）
df['buildUpPlaySpeedClass'].nunique()
df['buildUpPlaySpeedClass'].unique()
# buildUpPlaySpeedClass 唯一值分布（计数）
df['buildUpPlaySpeedClass'].value_counts()

# buildUpPlaySpeedClass 为 Fast 的球队名称（需关联 team 表）
team = pd.read_csv('../data/team.csv')
fast_teams = df[df['buildUpPlaySpeedClass'] == 'Fast'].merge(team, on='team_api_id')
print(fast_teams[['team_long_name', 'team_short_name']])
```
