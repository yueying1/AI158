```python
import numpy as np
import pandas as pd
```

```python
# >加载比赛数据集
df = pd.read_csv('..\data\match.csv')
df.head(3)
# >实现数据基本信息检视
df.describe()
df.info()
# >通过isnull0判断各列是否存在空值
df.isnull().sum()
# >以默认方式删除数据集的缺失值
df.dropna(inplace=True)
df.isnull().sum()
# >显示数据的描述信息
df.info()
# >使用value_counts(检测主队进球和客队进球的异常值
df.home_team_goal.value_counts()
df.away_team_goal.value_counts()
# >过滤异常数据
df = df[(df.home_team_goal>=0) & (df.away_team_goal>=0)]
df.info()
# >判断重复数据(以所有字段均相同为判断重复的标准
df.duplicated().sum()
```
