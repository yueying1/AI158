```python
import numpy as np
import pandas as pd
```

```python
# 周考
# 第一题:
#读取表格 input.csv请完成下面题目，并添加注释说明。（pandas实现）
df = pd.read_csv('input.csv')
# 1. 请打印出，字段”Field_01“，每种值出现的次数。
df['Field_01'].value_counts()
# 2. 清洗”Field_02“，仅保留汉字，删除所有字母、数字于空白符（包含换行符）。
df['Field_02'] = df['Field_02'].str.replace(r'[^\u4e00-\u9fff]', '', regex=True)
# 3. 清洗”Field_03“，仅保留数字。
df['Field_03'] = df['Field_03'].str.replace(r'\D', '', regex=True)
# 4. 给”Field_05“赋值，值为”Field_04“中”s“出现的次数。
df['Field_05'] = df['Field_04'].str.count('s')
# 5. 删除序号为 4 的一行。
df.drop(df[df['序号'] == 4].index)  #, inplace=True)
# 6. 填充”Field_06“列中的空值为”W“。将最终结果保存为同级目录下的 input_r.csv
df['Field_06'].fillna('W')  #, inplace=True)
df.to_csv('input_r.csv', index=False)
```

```python
# 第二题:
# #读取表格 超市.xls请完成下面题目
df = pd.read_excel('超市.xls')
#1. 从 DataFrame 中选择 订单日期、销售额 和 利润 这三列。
df[['订单日期', '销售额', '利润']]
#2. 检查数据中'销售额'是否存在缺失值，如果有，将 销售额列的缺失值填充为 0。
# df['销售额'].isnull().any()
if df['销售额'].isnull().any():
    df['销售额'].fillna(0, inplace=True)
    print('销售额中的缺失值已填充0')
else:
    print('销售额没有缺失值')
#3. 检查数据中'数量'是否存在缺失值，如果有，将数量 列的缺失值填充为 0。
if df['数量'].isnull().any():
    df['数量'].fillna(0, inplace=True)
    print('数量中的缺失值已填充0')
else:
    print('数量没有缺失值')
#4. 按照 销售额 降序对数据进行排序。
df.sort_values('销售额', ascending=False)
#5. 按照 利润 降序对数据进行排序。
df.sort_values('利润', ascending=False)
#6. 筛选出 销售额 大于 100 且 利润 为正数的订单。
df.query('销售额>100 and 利润>0')
#7. 将 省/自治区 列重命名为 省份。
df.rename(columns={'省/自治区': '省份'})
#8. 将 装运模式 列中的 二级 替换为 二级配送。
df['装运模式'].replace('二级', '二级配送', inplace=True)
#9. 计算 销售额、数量 和 利润 之间的相关性矩阵。
df[['销售额', '数量', '利润']].corr()
#10. 筛选出 销售额 大于其均值加 3 倍标准差的订单数据。
df[df['销售额'] > df['销售额'].mean() + 3 * np.std(df['销售额'])]
#11. 筛选出 类别 为 “办公用品” 且 区域 为 “华东” 的订单，并计算这些订单的总 数量。
mask = df.query('类别=="办公用品" and 区域=="华东"')
mask['数量'].sum()
```
