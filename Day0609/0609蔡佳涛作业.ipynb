```python
import numpy as np
import pandas as pd
```

```python
# 对比赛表match表的数据处理操作
df = pd.read_csv('../data/match.csv')
df.head(2)
# 将赛季字段的/更换为-(请尝试使用多种方法)
df['season'].str.replace('/','-')
df['season'].apply(lambda x:x.replace('/','-'))
df.replace('/','-',regex=True)
# 对主队进球列实现等宽分箱操作，分为5个区间，并实现分箱后值的统计
pd.cut(df['home_team_goal'],bins=5)
pd.cut(df['home_team_goal'],bins=5).value_counts().sort_index()
# 对客队进球列实现等宽分箱操作，分为5个区间，并实现分箱后值的统计
pd.cut(df['away_team_goal'],bins=5)
pd.cut(df['away_team_goal'],bins=5).value_counts().sort_index()
# 统计每条比赛记录已经距现在多少年，并将获得的数据设置为matchyear
pd.to_datetime(df['date'])
today_year = pd.to_datetime('today').year
df.assign(
    matchyear = df['date'].apply(lambda x:today_year-pd.to_datetime(x).year)
)
# df.assign(
#     matchyear = pd.to_datetime(df['date']).apply(lambda x:today_year-pd.to_datetime(x).year)
# )
# 设置比赛日期date为索引
df.date = pd.to_datetime(df.date)
df.set_index('date',inplace=True)
df
# 获得2008-08-17的全部比赛数据
df.loc['2008-08-17']
# 获得2008-08的全部比赛数据
df.loc['2008-08']
# 利用时间索引获得2008-08-17到2008-08-31的全部比赛数据
df.loc['2008-08-17':'2008-08-31']
```
