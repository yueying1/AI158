```python
def median(list, low, high):
    i = low - 1
    pivot = list[high]
    for j in range(low, high):
        if list[j] <= pivot :
            i += 1
            list[i], list[j] = list[j], list[i]
    list[i + 1], list[high] = list[high], list[i + 1]
    return i + 1
def quickSort(list, low, high):
    if low < high :
        medianNum = median(list, low, high)
        quickSort(list,low,medianNum - 1)
        quickSort(list,medianNum + 1, high)
list_ = [8, 15, 4, 17, 7, 19, 10, 3, 6, 9]
n = len(list_)
quickSort(list_, 0 ,n -1 )
print(list_)
```

```python
list1=[8,2,16,7,3,-8]
def q_fun(arr):
    if len(arr) <= 1 :
        return arr
    pivot = arr[len(arr) // 2]
    leftArr = [x  for x in arr if x < pivot]
    rightArr = [x  for x in arr if x > pivot]
    middleArr = [x for x in arr if x == pivot]
    return q_fun(leftArr) + middleArr + q_fun(rightArr)
print(q_fun(list1))
```
