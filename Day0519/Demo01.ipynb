```python
def printinfo( arg1, **vardict ):
   "打印任何传入的参数"
   print ("输出: ")
   print (arg1)
   print (vardict)

# 调用printinfo 函数
printinfo(1, a=2,b=3) # 1 {'a': 2, 'b': 3}
```

```python
def sum_numbers(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

result = sum_numbers(1, 2, 3, 4, 5)
print(result)  # 输出: 15
```
