```python
import numpy as np
import pandas as pd
player1 = pd.Series([98, 69, 73, 88, 100, 99],index=['体能','平衡','进攻','速度','射门','头球'])
player2 = pd.Series([71, 94, 99, 92, 76, 89],index=['体能','平衡','进攻','速度','射门','头球'])

# 获取某个球员评分Series的索引以及元素的个数
print('获取某个球员评分Series的索引以及元素的个数')
print(f'索引为：{player2.index},元素个数为：{player2.size}')
# 实现两个球员各项属性平均分的获取
print('现两个球员各项属性平均分的获取')
print((player1 + player2) / 2)
# 实现某个球员体能、进攻评分的获取
print('实现某个球员体能、进攻评分的获取')
print(f'球员1：体能：{player1.iloc[0]}，进攻：{player1.iloc[2]}')
print(f'球员2：体能：{player2.iloc[0]}，进攻：{player2.iloc[2]}')
# 实现每个球员超过95分的属性和评分
print('每个球员超过95分的属性和评分')
print(player1[player1 > 95])
print(player1[player2 > 95])
# 获取每个球员后三个属性的评分
print('每个球员后三个属性的评分')
print(player1[-3:])
print(player2[-3:])
# 获取球员1比球员2评分高的属性以及评分
print('球员1比球员2评分高的属性以及评分')
print(player1[player1 > player2])
```

```python
highs = pd.Series(['188cm','180cm','178cm','189cm','191cm','180cm',np.NAN], index=['球员1','球员2','球员3','球员4','球员5','球员6','球员7'])
# highs
# 统计不为空的球员个数
print('统计不为空的球员个数')
print(highs.notnull().sum())
# 处理身高数据去掉cm
print('处理身高数据去掉cm')
print(highs.str.replace('cm', ''))
# 统计180身高的球员个数
print('统计180身高的球员个数')
print(highs[highs == '180cm'].count())
# 处理身高数据为数值类型
print('处理身高数据为数值类型')
highs = highs.str.replace('cm', '').astype('float')
print(highs)
# 统计185以上身高的球员个数
print('统计185以上身高的球员个数')
print(highs[highs > 185].count())
```
