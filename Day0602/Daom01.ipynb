```python
import numpy as np
import pandas as pd
import matplotlib
import seaborn
from matplotlib.pyplot import title

np.__version__
pd.__version__
matplotlib.__version__
seaborn.__version__
```

```python
# python代码
def python_sum(n):
    a = [i**2 for i in range(n)]
    b = [i**3 for i in range(n)]
    return [a[i]+b[i] for i in range(n)]
    # return a + b
```

```python
# numpy代码
def numpy_sum(n):
    a = np.arange(n) ** 2
    b = np.arange(n) ** 3
    return  a + b
```

```python
# 一万次
%timeit python_sum(10000)
%timeit numpy_sum(10000)
```

```python
# 十万次
%timeit python_sum(100000)
%timeit numpy_sum(100000)
```

```python
# 百万次
%timeit python_sum(1000000)
%timeit numpy_sum(1000000)
```

```python
# 把上边的数据做成表
pd.DataFrame({
    'python':[2.4*1000,26.6*1000,274*1000],
    'numpy':[31.6,270,5.39*1000]
},index=['一万','十万','百万'])
```

```python
# 把上边表做成图
pd.DataFrame({
    'python':[2.4*1000,26.6*1000,274*1000],
    'numpy':[31.6,270,5.39*1000]
},index=['一万','十万','百万']).plot.bar(title = 'python和numpy的计算性能对比')
```
