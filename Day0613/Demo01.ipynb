```python
# 需求说明
# 全部数据评分与潜能的关系
# 每个球员平均评分与平均潜能的关系
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
attr = pd.read_csv('../data/player_attributes.csv')

# ========== 1. 全部数据：评分与潜能的关系 ==========
# 相关性
print('全部数据评分与潜能的相关度：')
print(attr['overall_rating'].corr(attr['potential']))

# 散点图
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(attr['overall_rating'], attr['potential'], alpha=0.3, s=5)
plt.xlabel('评分')
plt.ylabel('潜能')
plt.title('全部数据：评分 vs 潜能')

# ========== 2. 每个球员平均评分与平均潜能的关系 ==========
player_avg = attr.groupby('player_api_id')[['overall_rating', 'potential']].mean()

print('\n每个球员平均评分与平均潜能的相关度：')
print(player_avg['overall_rating'].corr(player_avg['potential']))

plt.subplot(1, 2, 2)
plt.scatter(player_avg['overall_rating'], player_avg['potential'], alpha=0.4, s=10)
plt.xlabel('平均评分')
plt.ylabel('平均潜能')
plt.title('球员平均值：评分 vs 潜能')

plt.tight_layout()
plt.show()
```
